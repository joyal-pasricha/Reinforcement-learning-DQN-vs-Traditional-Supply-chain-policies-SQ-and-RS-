Comparing the 3 policies(SQ, RS AND DQN i.e. RL) using graphs(results on test data) -

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dqn_data = pd.read_excel("/content/DQN_test_data_results.xlsx")
sq_data = pd.read_excel("/content/SQ_test_data_results.xlsx")
rs_data = pd.read_excel("/content/RS_test_data_results.xlsx")
# Filter data for the first 12 episodes
dqn_filtered = dqn_data[dqn_data['episode'] < 12].groupby('episode').mean()
sq_filtered = sq_data[sq_data['episode'] < 12].groupby('episode').mean()
rs_filtered = rs_data[rs_data['episode'] < 12].groupby('episode').mean()

# Plot 1: Total Costs Comparison
plt.figure(figsize=(10, 6))
plt.plot(dqn_filtered.index, dqn_filtered['total_cost'], label='DQN')
plt.plot(sq_filtered.index, sq_filtered['total_cost'], label='SQ')
plt.plot(rs_filtered.index, rs_filtered['total_cost'], label='RS')
plt.title('Total Costs Comparison')
plt.xlabel('Episode')
plt.ylabel('Total Cost')
plt.legend()
plt.grid()
plt.show()

# Plot 2: Number of Wasted Units Comparison
plt.figure(figsize=(10, 6))
plt.plot(dqn_filtered.index, dqn_filtered['wasted_units'], label='DQN')
plt.plot(sq_filtered.index, sq_filtered['wasted_units'], label='SQ')
plt.plot(rs_filtered.index, rs_filtered['wasted_units'], label='RS')
plt.title('Number of Wasted Units Comparison')
plt.xlabel('Episode')
plt.ylabel('Wasted Units')
plt.legend()
plt.grid()
plt.show()

# Plot 3: Orders Comparison
plt.figure(figsize=(10, 6))
plt.plot(dqn_filtered.index, dqn_filtered['order_quantity'], label='DQN')
plt.plot(sq_filtered.index, sq_filtered['order_quantity'], label='SQ')
plt.plot(rs_filtered.index, rs_filtered['order_quantity'], label='RS')
plt.title('Orders Comparison')
plt.xlabel('Episode')
plt.ylabel('Order Quantity')
plt.legend()
plt.grid()
plt.show()

# Plot 4: Unfulfilled Demand or lost sales comparison
plt.figure(figsize=(10, 6))
plt.plot(dqn_filtered.index, dqn_filtered['unfulfilled_demand'], label='DQN Unfulfilled Demand')
plt.plot(sq_filtered.index, sq_filtered['unfulfilled_demand'], label='SQ Unfulfilled Demand')
plt.plot(rs_filtered.index, rs_filtered['unfulfilled_demand'], label='RS Unfulfilled Demand')
plt.title('Unfulfilled demand or lost sales comparison')
plt.xlabel('Episode')
plt.ylabel('Units')
plt.legend()
plt.grid()
plt.show()

# Plot 5: Fill Rate Comparison
plt.figure(figsize=(10, 6))
plt.plot(dqn_filtered.index, dqn_filtered['fill_rate'], label='DQN')
plt.plot(sq_filtered.index, sq_filtered['fill_rate'], label='SQ')
plt.plot(rs_filtered.index, rs_filtered['fill_rate'], label='RS')
plt.title('Fill Rate Comparison')
plt.xlabel('Episode')
plt.ylabel('Fill Rate')
plt.legend()
plt.grid()
plt.show()

# Plot 6: Service Cycle Rate Comparison
plt.figure(figsize=(10, 6))
plt.plot(dqn_filtered.index, dqn_filtered['service_cycle_rate'], label='DQN')
plt.plot(sq_filtered.index, sq_filtered['service_cycle_rate'], label='SQ')
plt.plot(rs_filtered.index, rs_filtered['service_cycle_rate'], label='RS')
plt.title('Service Cycle Rate Comparison')
plt.xlabel('Episode')
plt.ylabel('Service Cycle Rate')
plt.legend()
plt.grid()
plt.show()

# Plot 7: Heatmaps
plt.figure(figsize=(12, 6))
heatmap_data = pd.concat([
    dqn_filtered[['total_cost', 'wasted_units', 'fill_rate', 'service_cycle_rate']].rename(columns=lambda x: f'DQN_{x}'),
    sq_filtered[['total_cost', 'wasted_units', 'fill_rate', 'service_cycle_rate']].rename(columns=lambda x: f'SQ_{x}'),
    rs_filtered[['total_cost', 'wasted_units', 'fill_rate', 'service_cycle_rate']].rename(columns=lambda x: f'RS_{x}')
], axis=1)
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Comparison Between Policies')
plt.show()
```
